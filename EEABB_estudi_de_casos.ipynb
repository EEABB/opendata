```python
import requests
import json

key = 'EL_TOKEN_QUE_ENS_DONA_METEOCAT';


def detalleMunicipio(variableTemperatura, codigoPoblacion):

  url = 'https://api.meteo.cat/xema/v1/representatives/metadades/municipis/'+codigoPoblacion+'/variables/'+str(variableTemperatura)
  response = requests.get(url, headers={"Content-Type": "application/json", "X-Api-Key": key})
  detalleMunicipio = json.loads(response.text)
  return detalleMunicipio



def getMunicipio(nombre_municipio_buscar):
  urlMunicipis = 'https://api.meteo.cat/referencia/v1/municipis'
  response = requests.get(urlMunicipis, headers={"Content-Type": "application/json", "X-Api-Key": key})
  municipios = json.loads(response.text)

  municipio_encontrado = next((m for m in municipios if m["nom"] == nombre_municipio_buscar), None)
  return municipio_encontrado

municipio = getMunicipio("Sallent")

#variableTemperatura = 32
#codigoPoblacion = "081385"

variableTemperatura = 32
codigoPoblacion = municipio['codi']

detalle = detalleMunicipio(variableTemperatura, codigoPoblacion)
print(detalle);


```
