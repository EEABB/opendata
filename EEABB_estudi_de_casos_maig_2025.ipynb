```python
import requests
import pandas as pd
from pandas import json_normalize


url = 'https://males-herbes-api.aaaida.com/weatherDatas'

response = requests.get(url)
data = response.json()

print(data)
# Crear el DataFrame con los datos
df = pd.DataFrame(data['result'])
print(df.info())


expanded_weatherStation = json_normalize(df['weatherStation'])
df = pd.concat([df, expanded_weatherStation], axis=1)
df.drop(columns=['weatherStation'], inplace=True)

expanded_data = json_normalize(df['data'])
df = pd.concat([df, expanded_data], axis=1)
df.drop(columns=['data'], inplace=True)

max = df["avgAirTemperature"].max()
min = df["avgAirTemperature"].min()
mean = df["avgAirTemperature"].mean()
median = df["avgAirTemperature"].median()
mode = df["avgAirTemperature"].mode()[0]

print("=============")
print(max)
print(min)
print(mean)
print(median)
print(mode)
print("=============")

# Mostrar el DataFrame
print(df.info())
print(df.loc[0])
print("===========")
#print(df.corr())
```

```python
import requests
import pandas as pd
from pandas import json_normalize



url = 'https://analisi.transparenciacatalunya.cat/resource/gn9e-3qhr.json'

response = requests.get(url)
data = response.json()

print(data)
# Crear el DataFrame con los datos
df = pd.DataFrame(data)
print(df.info())
# Conversión de tipos
df['volum_embassat'] = df['volum_embassat'].astype(float)
df['percentatge_volum_embassat'] = df['percentatge_volum_embassat'].astype(float)
df['nivell_absolut'] = df['nivell_absolut'].astype(float)
df['dia'] = pd.to_datetime(df['dia'])

# Bar chart: Volumen por embalse
fig = px.bar(df, x='estaci', y='volum_embassat', color='percentatge_volum_embassat',
             title='Volumen embalsado por embalse',
             labels={'estaci': 'Embalse', 'volum_embassat': 'Volumen (hm³)'})
fig.show()
```

```python
import plotly.express as px

df = pd.DataFrame(data)

df['percentatge_volum_embassat'] = df['percentatge_volum_embassat'].astype(float)

df = df.sort_values(by='percentatge_volum_embassat', ascending=True)  # para barras horizontales de menor a mayor


fig = px.bar(
    df,
    x='percentatge_volum_embassat',
    y='estaci',
    orientation='h',
    title='Ranking per percentatge de volum embassat',
    labels={'percentatge_volum_embassat': '% volum embassat', 'estaci': 'Embassament'},
    text='percentatge_volum_embassat',
    color='percentatge_volum_embassat',
    color_continuous_scale='Blues'
)


fig.update_traces(texttemplate='%{text:.1f}%', textposition='outside')
fig.update_layout(xaxis_title='% de volum embassat', yaxis_title='Embassament', yaxis=dict(tickfont=dict(size=10)))

fig.show()
```
